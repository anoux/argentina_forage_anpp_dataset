```python
#delete some columns not usefull for this data exploration and visualizations

df = df.drop(columns = [
 'DesvEst interanual (n)',
 'DesvEst interanual (kg/ha)',
 'DesvEst espacial (kg/ha)',
 'Edad (años)',
 'Fertilización  (S) (kg/ha.año)',
 'Fertilización  (K) (kg/ha.año)',
 'Pastoreo',
 'Clausura',
 'Precipitación (mm)',
 'Precipitación durante el año de medición (mm/estación de crecimiento)',
 'Precipitación promedio mencionada en la fuente (mm)',
 'Precip. anual promedio según SMN interpolado (mm/año)',
 'Código de Unidad de Vegetación',
 'Nombre fisionómico de Unidad de Vegetación',
 'Nombre vulgar de Unidad de Vegetación',
 ])
```

```python
# Columns renaming. Headers translation to English

df.rename(columns={'ID Registro':'Record ID',
 'ID fuente': 'Source ID',
 'Celda':'Cell',
 'Departamento, partido o comuna':'Department, district, or commune',
 'Provincia':'Province',
 'Espontánea':'Spontaneous vegetation',
 'Recurso 1':'Resource type 1',
 'Recurso 2':'Resource type 2',
 'Fuente de variación':'Variability sources',
 'Sitioy/o condición':'About site, condition',
 'Composición florística':'Floristic Composition',
 'Composición florística simplificada (solo vegetación cultivada)':'Simplified Floristic Composition',
 'Riqueza especies':'Species richness',
 'PPNA (kg/ha.año)':'ANPP (kg/ha.year)',
 'Período evaluado (años)':'Evaluation period',
 'Inicio':'Start',
 'Final':'End',
 'Cantidad de años evaluados':'Years evalauted',
 'Fertilización':'Fertilization',
 'Tratamiento':'Treatment',
 'Fertilización (N) (kg/ha.año)':'N fertilization (kg/ha.year)',
 'Fertilización  (P) (kg/ha.año)':'P fertilization (kg/ha.year)',
 'Riego':'Irrigation',
 'Cantidad riego (mm)':'Irrigation milimeters (mm)',
 'Precipitación media anual tomada de geoportal IDESA':'Mean annual precipitation from IDESA web',
 'Taxonomía de suelo':'Soil taxonomy',
 'Metodología/Comentarios':'Metodology and comments',
 'REGION':'Major Region',
 'DOMINIO':'Domain',
 'UNIDAD FITOGEOGRÁFICA':'Region',
 'Excluidos del análisis':'Exlcuded from analysis'}, inplace=True)
```

```python
df
```

```python
df['Exlcuded from analysis'].unique()

# I will drop the rows which have not 'nan' for this column since they correspond
# to rare anpp values
```

```python
# I want to exclude here the records (rows) that were excluded from the analysis I already did
# That means those that are not NaN for the "Excluidos del análisis" column

df['Exlcuded from analysis'] = df['Exlcuded from analysis'].fillna('OK')

# First fill de NaN cells with 'OK'
# Now I will drop the rows that are different from 'OK' for that column

for x in df.index:
    if df.loc[x, "Exlcuded from analysis"] != 'OK':
        df.drop(x, inplace=True)
        
df
```

```python
# Run this again to check if the excluded rows where actually excluded

df['Exlcuded from analysis'].unique()

# Great
```

```python
df_by_resource_type = df.groupby('Resource type 2')['ANPP (kg/ha.year)'].mean().reset_index()

# The reset.index() function converts 
# right away the group by in a new dataframe

df_by_resource_type
```

```python
# Rename headers for plot

df_by_resource_type = df_by_resource_type.rename(columns={
    'Resource type 2': 'Resource', 
    'ANPP (kg/ha.year)': 'mean ANPP (kg/ha.year)'})

df_by_resource_type
```

```python
df_by_resource_type.drop(df_by_resource_type.tail(1).index,inplace=True)

# Exlcuded the "otro" resource since it is a weird case
```

```python
df_by_resource_type
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab


fig_df_by_resource_type = df_by_resource_type.set_index('Resource')

# So the names for each colmun along the x axis are 
#the different resources instead on numbers

fig_df_by_resource_type.plot(kind = 'bar', title = 'Mean ANPP by resource',
                        xlabel = 'Resource', 
                        ylabel = 'Mean ANPP (kg/ha.year-1)',
                        rot=0,
                        figsize = (10,4),
                        legend = False)


#Font size setting for plots
 
SMALL_SIZE = 5
MEDIUM_SIZE = 10
BIGGER_SIZE = 20

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
# I would like to see how ANPP differs between fitogeographic regions instead of resource type

df_by_region = df.groupby('Region')['ANPP (kg/ha.year)'].mean().reset_index()
```

```python
# Rename the headers for plot

df_by_region = df_by_region.rename(columns={'PPNA (kg/ha.año)': 'mean ANPP (kg/ha.year)'})
```

```python
# Not all regions are well represented, some of them have just a few records, 
#such as the Ecotono Monte-Patagonia and the Punena region

df.groupby('Region')['ANPP (kg/ha.year)'].count()
```

```python
# Since some regions have just a few records I will exclude them of the next plot

poor_data_regions = [0, 1, 4, 6, 7, 8, 9]

df_by_region.drop(poor_data_regions, axis=0, inplace=True)
```

```python
df_by_region = df_by_region.reset_index(drop=True)

df_by_region
```

```python
df_by_region['Region'] = df_by_region['Region'].str.replace('Provincia Fitogeografica ','')

df_by_region
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab


fig_df_by_region = df_by_region.set_index('Region')

# So the names for each colmun along the x axis are 
# the different regions instead on numbers

fig_df_by_region.plot(kind = 'bar', title = 'Mean ANPP by region',
                        xlabel = 'Region', 
                        ylabel = 'Mean ANPP (kg/ha.year-1)',
                        rot=0,
                        figsize = (10,4),
                        legend = False,
                        color = "green")


#Font size setting for plots
 
SMALL_SIZE = 5
MEDIUM_SIZE = 10
BIGGER_SIZE = 20

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```
