```python
my_list = [2547.658114, 9012.541957, 6402.379562, 15254.226219]
my_index = ['Grassland', 'Pasture', 'Winter crop', 'Summer crop']


my_series = pd.Series(my_list, index=my_index)
```

```python
df_by_resource_type = pd.DataFrame({'Resource': my_series.index, 'mean ANPP': my_series.values})
```

```python
df_by_resource_type
```

```python
df_by_resource_type.sort_values(by="mean ANPP",ascending=False)

# Summer short term crops are the most productive ones, then pastures,
# the winter short term crops, and last of all grasslands
```

```python
df_by_resource_type = df_by_resource_type.reset_index(drop=True)

df_by_resource_type
```

```python
fig_df_by_resource_type = df_by_resource_type.set_index('Resource')

# So the names for each colmun along the x axis are the different resources instead on numbers

fig_df_by_resource_type.plot(kind = 'bar', title = 'mean ANPP by resource',
                        xlabel = 'Resource', ylabel = 'Mean ANPP (kg/ha.year-1)',
                        rot=0, legend = False)
```
